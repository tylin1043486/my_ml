### 資料載入及套件載入

```python
from sklearn.datasets import fetch_olivetti_faces
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.utils.validation import check_random_state
%matplotlib inline
```

```python
face_row_data = fetch_olivetti_faces()
X = face_row_data.data
y = face_row_data.target
```

### 資料分割

```python
#print(X.shape,y.shape)
#print(y)
np.sqrt(X.shape[1]) ##每筆資料像素為64x64
##將資料整理成3維
X_3d = X.reshape(400,64,64)

x_train = X_3d[0:300] #實際取第0筆到299筆
print(x_train.shape)
y_train = y[0:300]
#print(y_train)
x_test = X_3d[300:] #取第300筆(含)以後的資料
#print(x_test.shape)
y_test = y[300:]
#print(y_test)

#這個cell的y對整個資料輸出沒有幫助
```

### 看一下資料的樣子

```python
plt.figure(figsize=(10,5))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.axis('off')
    plt.imshow(x_train[i],cmap='gray')
```

```python
# Test on a subset of people
n_faces = 6
rng = check_random_state(2)
face_ids = rng.randint(x_test.shape[0], size=(n_faces, ))
test = x_test[face_ids, :]
print(test.shape)
```

### 將訓練資料分成上半臉(input)及下半臉(解答)

```python
x_train_upper = x_train[:,:32,:] ##上半臉為32x64 索引為0~31
y_train_lower = x_train[:,32:,:] ##答案
x_test_upper = test[:,:32,:]
y_test_lower = test[:,32:,:] ##正解
print(x_test_upper.shape)
```

### 看一下切割出來的圖片有沒有問題

```python
plt.figure(figsize=(10,4))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.axis('off')
    plt.imshow(x_train_upper[i],cmap='gray')
    #plt.imshow(y_train_lower[i])
    
#for i in range(n_faces):
    #plt.subplot(1,5,i+1)
    #plt.axis('off')
    #plt.imshow(x_test_upper[i])
    #plt.imshow(y_test_lower[i])
```

### 模型選擇

資料是三維的，而機器學習演算法能使用的維度為二維，所以資料最後還是需要轉回二維，神經網路應該可以使用多維度的資料

```python
x_train_upper = x_train_upper.reshape(300,-1)
y_train_lower = y_train_lower.reshape(300,-1)
x_test_upper = x_test_upper.reshape(test.shape[0],-1)
print(x_train_upper.shape,y_train_lower.shape,x_test_upper.shape)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import Pipeline
```

```python
# Fit estimators
ESTIMATORS = {
    "Extra trees": ExtraTreesRegressor(n_estimators=10, max_features=32,
                                       random_state=0),
    "K-nn": KNeighborsRegressor(),
    "Linear regression": LinearRegression(),
    "Ridge": RidgeCV(),
    'Decision tree' :DecisionTreeRegressor(max_depth = 5)
}

# 訓練
y_test_predict = {}
for name, estimator in ESTIMATORS.items():
    estimator.fit(x_train_upper, y_train_lower)
    y_test_predict[name] = estimator.predict(x_test_upper)
```

```python
# Plot the completed faces
image_shape = (64, 64)

n_cols = 1 + len(ESTIMATORS)
plt.figure(figsize=(2. * n_cols, 2.26 * n_faces))
plt.suptitle("Face completion with multi-output estimators", size=16)

for i in range(n_faces):
    true_face = test[i]
    if i !=0: # i <> 0
        sub = plt.subplot(n_faces, n_cols, i * n_cols + 1)
    else: # i=0 時，加標題
        sub = plt.subplot(n_faces, n_cols, i * n_cols + 1,
                          title="true faces")
    # 不顯示座標軸
    sub.axis("off")
    sub.imshow(true_face.reshape(image_shape),
               cmap=plt.cm.gray,
               interpolation="nearest")

    # 顯示各種演算法的預測結果
    for j, est in enumerate(sorted(ESTIMATORS)):
        # 上、下半部的臉接合
        completed_face = np.hstack((x_test_upper[i], y_test_predict[est][i]))

        if i: # i <> 0
            sub = plt.subplot(n_faces, n_cols, i * n_cols + 2 + j)
        else: # i=0 時，加標題
            sub = plt.subplot(n_faces, n_cols, i * n_cols + 2 + j,
                              title=est)

        sub.axis("off")
        sub.imshow(completed_face.reshape(image_shape),
                   cmap=plt.cm.gray,
                   interpolation="nearest")

plt.show()
```

* sixth row 鼻子的長度比資料庫的資料還長，整個就毀了
+ [6,4]圖直接變黑人的輪廓
+ 線性回歸出來的效果，肉眼看起來，最差

### 決策樹max_depth不同的效果

```python
depth_list = [1,5,15]
plt.figure(figsize=(2. * len(depth_list), 2.26 * n_faces))

for index,depth in enumerate(depth_list):
    tree = DecisionTreeRegressor(max_depth = depth)
    tree.fit(x_train_upper, y_train_lower)
    tree_predit=tree.predict(x_test_upper)
    
    for i in range(n_faces):
        #上下臉結合
        tree_face = np.hstack((x_test_upper[i],tree_predit[i]))
        if i:
            pic = plt.subplot(n_faces,len(depth_list),i*3+1+index)
        else:
            pic = plt.subplot(n_faces,len(depth_list),i*3+1+index,title='max_depth: %d' %depth)
        pic.axis('off')
        pic.imshow(tree_face.reshape(image_shape),
                    cmap=plt.cm.gray,
                    interpolation="nearest")
plt.show()
```

### 加上true_face

```python
depth_list = [1,5,15]
plt.figure(figsize=(2. * (len(depth_list)+1), 2.26 * n_faces))
for i in range(n_faces): 
    true_face = test[i]
    if i !=0: # i <> 0
        pic = plt.subplot(n_faces,len(depth_list)+1,i*4+1)
    else: # i=0 時，加標題
        pic = plt.subplot(n_faces,len(depth_list)+1,i*4+1,
                                  title="true faces")
     # 不顯示座標軸
    pic.axis("off")
    pic.imshow(true_face.reshape(image_shape),
                cmap=plt.cm.gray,
                interpolation="nearest")
for index,depth in enumerate(depth_list):
    tree = DecisionTreeRegressor(max_depth = depth)
    tree.fit(x_train_upper, y_train_lower)
    tree_predit=tree.predict(x_test_upper)
    
    for i in range(n_faces):
        #上下臉結合
        tree_face = np.hstack((x_test_upper[i],tree_predit[i]))
        if i:
            pic = plt.subplot(n_faces,len(depth_list)+1,i*4+2+index)
        else:
            pic = plt.subplot(n_faces,len(depth_list)+1,i*4+2+index,title='max_depth: %d' %depth)
        pic.axis('off')
        pic.imshow(tree_face.reshape(image_shape),
                    cmap=plt.cm.gray,
                    interpolation="nearest")
plt.show()
```
